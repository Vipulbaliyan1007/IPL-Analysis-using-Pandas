# Destructve death over batsman in the history of IPL ( isin() function)

```python
death_over = data1[data1['over']>15]
```

```python
all_batsman= death_over.groupby('batsman')['batsman_runs'].count().sort_values(ascending= False)
```

```python
x= all_batsman>200
batsman_list= all_batsman[x].index.tolist()
```

```python
final= data1[data1['batsman'].isin(batsman_list)] 
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls=final.groupby('batsman')['batsman_runs'].count()
```

```python
sr=(runs/balls)*100
```

```python
# These the the batsman with best strike rates 
sr.sort_values(ascending= False)
```
